## Exploratory Data Analysis

```python


merchant_columns = [ 'MERCHANT CATEGORY']
revenue_columns = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']


df_melted = pd.melt(df, id_vars=merchant_columns, value_vars=revenue_columns, var_name='Month', value_name='Revenue')

# Plotting a line chart
plt.figure(figsize=(12, 8))
sns.lineplot(x='Month', y='Revenue', hue='MERCHANT CATEGORY', data=df_melted, marker='o')
plt.title('Monthly Revenue Trends by Merchant')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend(title='MERCHANT CATEGORY', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()


```

```python


merchant_category = 'TEXTILE'
revenue_columns = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']

# Filter the DataFrame for the specific merchant category
df_textile = df[df['MERCHANT CATEGORY'] == merchant_category]

# Melt the filtered DataFrame
df_melted = pd.melt(df_textile, id_vars=['MERCHANT CATEGORY'], value_vars=revenue_columns, var_name='Month', value_name='Revenue')

# Plotting a line chart
plt.figure(figsize=(12, 8))
sns.lineplot(x='Month', y='Revenue', data=df_melted, marker='o')
plt.title(f'Monthly Revenue Trends for {merchant_category}')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend(title='MERCHANT CATEGORY', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

### Correlation Matrix - Revenue by Month and Merchant Category

```python

revenue_columns = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
selected_columns = ['MERCHANT CATEGORY'] + revenue_columns


df_selected = df[selected_columns]


correlation_matrix = df_selected.corr()


plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix - Revenue by Month and Merchant Category')
plt.show()

```

### Total Annual Revenue

```python


revenue_columns = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']


df['ANNUAL REVENUE'] = df[revenue_columns].sum(axis=1)


print(df[[ 'MERCHANT CATEGORY', 'ANNUAL REVENUE']])
```

```python

df['PROFIT'] = df['ANNUAL REVENUE'] - df['ANNUAL COST THE SERVICE THE MERCHANT']
df['PROFIT MARGIN'] = (df['PROFIT'] / df['ANNUAL REVENUE']) * 100


print(df)
```
